```python
import pandas as pd

# 1. Завантажуємо дані
df = pd.read_csv("C:/Users/kondr/Downloads/dataset_2026-04-15T12_57_56.996577045Z_DEFAULT_INTEGRATION_SSSU_DF_POPULATION_DEATH_CAUSE_LATEST.csv")

# 2. Визначаємо колонки, які нам потрібні
# Ми залишаємо демографію, причину та роки
needed_columns = [
    'Показник', 
    'Територіальний розріз', 
    'Вік', 
    'Стать', 
    'Причина смерті', 
    '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021'
]

# 3. Фільтруємо лише потрібні колонки
df_clean = df[needed_columns].copy()

# 4. ВАЖЛИВО: Окремо виділяємо абсолютні числа (особи) 
# Це допоможе не змішувати їх з коефіцієнтами
df_people = df_clean[df_clean['Показник'] == 'Кількість померлих'].drop(columns=['Показник'])

# 5. Заповнюємо порожні значення нулями (Явищ не було)
years = ['2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021']
df_people[years] = df_people[years].fillna(0)

# Тепер у вас є df_people — чистий датасет для аналізу кількості людей
print(df_people.head())
```

```python
# Зберігаємо очищений датасет у новий файл
df_people.to_csv('ukraine_mortality_cleaned.csv', index=False, encoding='utf-8-sig')
```
